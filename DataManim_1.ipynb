```python
# 카드이용데이터 : https://www.kaggle.com/sakshigoyal7/credit-card-customers
DataUrl5 = 'https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv'
print("\n56. 데이터를 로드하고 데이터 행과 열의 갯수를 출력하라")
data = pd.read_csv(DataUrl5)
print("행:%d, 열:%d" % (data.shape[0], data.shape[1]))

print("\n57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라")
# Unknown : N   # Less than $40K : a    # $40K - $60K : b   # $60K - $80K : c   # $80K - $120K : d  # $120K + : e
ic_dict = {
    'Unknown': 'N',
    'Less than $40K': 'a',
    '$40K - $60K': 'b',
    '$60K - $80K': 'c',
    '$80K - $120K': 'd',
    '$120K +': 'e'
}
data2 = data.copy()
data2['newIncome'] = data2['Income_Category'].map(lambda x: ic_dict[x])
display(data2[['Income_Category', 'newIncome']])

print("\n58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라")
# Unknown : N   # Less than $40K : a    # $40K - $60K : b   # $60K - $80K : c   # $80K - $120K : d  # $120K + : e
def ic_apply(x):
    if x == 'Unknown': return 'N'
    if x == 'Less than $40K': return 'a'
    if x == '$40K - $60K': return 'b'
    if x == '$60K - $80K': return 'c'
    if x == '$80K - $120K': return 'd'
    if x == '$120K +': return 'e'
data2 = data.copy()
data2['newIncome'] = data2['Income_Category'].apply(ic_apply)
display(data2[['Income_Category', 'newIncome']])

print("\n59. Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라. (0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력하라")
def ages(x):
    if x >= 0 and x < 10: return 0
    elif x < 20: return 10
    elif x < 30: return 20
    elif x < 40: return 30
    elif x < 50: return 40
    elif x < 60: return 50
    elif x < 70: return 60
    else: return 70
data2 = data.copy()
data2['AgeState'] = data2['Customer_Age'].apply(ages)
display(data2[['Customer_Age', 'AgeState']])
display(data2['AgeState'].value_counts().sort_index().to_frame())

print("\n60. Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력하라")
import numpy as np

data2 = data.copy()
data2['newEduLevel'] = np.where(data2['Education_Level'].str.contains('Graduate'), 1, 0)
display(data2[['Education_Level', 'newEduLevel']])
display(data2['newEduLevel'].value_counts().sort_index().to_frame())
# data['newEduLevel'] = data.Education_Level.map(lambda x : 1 if 'Graduate' in x else 0)
```

## 66~70 (*67 datetime)

```python
print("66. Yr_Mo_Dy에 존재하는 년도의 유일값을 모두 출력하라")
print(data['Yr_Mo_Dy'].dt.year.unique())

print("\n67. Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라")
import datetime

def fix_year(x):
    Y = x.year
    if Y >= 2061:
        Y -= 100
    return pd.to_datetime(datetime.date(Y, x.month, x.day))
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_year)

print("\n68. 년도별 각컬럼의 평균값을 구하여라")
# data['Yr'] = data['Yr_Mo_Dy'].dt.year
# display(data.groupby('Yr').mean())
display(data.groupby(data.Yr_Mo_Dy.dt.year).mean())

print("\n69. weekday컬럼을 만들고 요일별로 매핑하라 ( 월요일: 0 ~ 일요일 :6)")
data['weekday'] = data['Yr_Mo_Dy'].dt.weekday
display(data[['Yr_Mo_Dy', 'weekday']])

print("\n70. weekday컬럼을 기준으로 주말이면 1 평일이면 0의 값을 가지는 WeekCheck 컬럼을 만들어라")
def week(x):
    if x >= 5: return 1
    else: return 0
data['WeekCheck'] = data['weekday'].apply(week)
display(data[['weekday', 'WeekCheck']])
```

## 71~75 (*72 fillna ffill/bfill, 74 diff 1차차분, 75 rolling 이동평균/합)

```python
print("71. 년도, 일자 상관없이 모든 컬럼의 각 달의 평균을 구하여라")
display(data.groupby(data['Yr_Mo_Dy'].dt.month).mean(numeric_only=True))

print("\n72. 모든 결측치는 컬럼기준 직전의 값으로 대체하고 첫번째 행에 결측치가 있을경우 뒤에있는 값으로 대채하라")
display(data.fillna(method='ffill').fillna(method='bfill'))

print("\n73. 년도 - 월을 기준으로 모든 컬럼의 평균값을 구하여라")
# display(data.groupby([data['Yr_Mo_Dy'].dt.year, data['Yr_Mo_Dy'].dt.month]).mean(numeric_only=True))
display(data.groupby(data['Yr_Mo_Dy'].dt.to_period('M')).mean())

print("\n74. RPT 컬럼의 값을 일자별 기준으로 1차차분하라")
display(data['RPT'].diff())

print("\n75. RPT와 VAL의 컬럼을 일주일 간격으로 각각 이동평균한값을 구하여라")
display(data[['RPT', 'VAL']].rolling(window=7).mean())
```

```python
# 서울시 미세먼지 데이터 : https://www.airkorea.or.kr/web/realSearch?pMENU_NO=97
DataUrl7 = 'https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv'
print("76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다")
data = pd.read_csv(DataUrl7)
def time024(x):
    date, hour = x.split(':')
    if hour =='24':
        hour ='00:00:00'
        FinalDate = pd.to_datetime(date + " " + hour) + datetime.timedelta(days=1)
    else:
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date + " " + hour)
    return FinalDate
data['(년-월-일:시)'] = data['(년-월-일:시)'].apply(time024)
display(data['(년-월-일:시)'])

print("\n77. 일자별 영어요일 이름을 dayName 컬럼에 저장하라")
data['dayName'] = data['(년-월-일:시)'].dt.weekday.map({0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'})
# data['dayName'] = data['(년-월-일:시)'].dt.day_name()
display(data[['(년-월-일:시)', 'dayName']])

print("\n78. 일자별 각 PM10등급의 빈도수를 파악하라")
display(data.groupby([data['(년-월-일:시)'].dt.to_period('D'), 'PM10등급']).size().unstack().fillna(0))

print("\n79. 시간이 연속적으로 존재하며 결측치가 없는지 확인하라")
if len(data['(년-월-일:시)'].diff().unique()) == 2: # 첫 행 NaN과 시간 1
    print("True")
else:
    print("False")

print("\n80. 오전 10시와 오후 10시(22시)의 PM10의 평균값을 각각 구하여라")
data10am = data.loc[data['(년-월-일:시)'].dt.hour == 10]
data10pm = data.loc[data['(년-월-일:시)'].dt.hour == 22]
print(data10am['PM10'].mean(), data10pm['PM10'].mean())
# data.groupby(df['(년-월-일:시)'].dt.hour).mean().iloc[[10,22],[0]]
```
