## 1~5 (*3)

```python
# 데이터 출처 :https://www.kaggle.com/rsrishav/youtube-trending-video-dataset?select=KR_youtube_trending_data.csv
# 데이터 설명 : 유튜브 데일리 인기동영상 (한국)
DataUrl1 = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv'
data = pd.read_csv(DataUrl1, index_col=0)
display(data)
```

```python
print("1. 인기동영상 제작횟수가 많은 채널 상위 10개명을 출력하라 (날짜기준, 중복포함)")
print(data.groupby(['channelTitle']).size().sort_values(ascending=False)[:10].index)

print("\n2. 논란으로 인기동영상이 된 케이스를 확인하고 싶다. dislikes수가 like 수보다 높은 동영상을 제작한 채널을 모두 출력하라")
print(data.loc[data['dislikes'] > data['likes']].groupby('channelTitle').size().index)

print("\n3. 채널명을 바꾼 케이스가 있는지 확인하고 싶다. channelId의 경우 고유값이므로 이를 통해 채널명을 한번이라도 바꾼 채널의 갯수를 구하여라")
data3 = data[['channelId', 'channelTitle']].drop_duplicates()['channelId'].value_counts()
print(len(data3.loc[data3 > 1]))

print("\n4. 일요일에 인기있었던 영상들중 가장많은 영상 종류(categoryId)는 무엇인가?")
data['trending_date2'] = pd.to_datetime(data['trending_date2'])
data4 = data.loc[data['trending_date2'].dt.weekday == 6]
print(data4.groupby('categoryId').size().sort_values().to_frame().index[-1])

print("\n5. 각 요일별 인기 영상들의 categoryId는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라")
display(data.pivot_table(index=data['trending_date2'].dt.weekday, columns='categoryId', aggfunc='size'))
```

## 6~10 (*9)

```python
print("6. 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다. viewcount대비 댓글수가 가장 높은 영상을 확인하라 (view_count값이 0인 경우는 제외한다)")
data['ratio'] = np.where(data['view_count'] > 0, data['comment_count'] / data['view_count'], np.nan)
print(data.loc[data['ratio'] == data['ratio'].max()]['title'].tolist()[0])

print("\n7. 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다.viewcount대비 댓글수가 가장 낮은 영상을 확인하라 (view_counts, ratio값이 0인경우는 제외한다.)")
print(data.loc[data['ratio'] > 0].dropna().sort_values(by='ratio', ascending=True)['title'].iloc[0])

print("\n8. like 대비 dislike의 수가 가장 적은 영상은 무엇인가? (like, dislike 값이 0인경우는 제외한다)")
data8 = data.copy()
data8['dis/like'] = np.where((data8['likes'] > 0) & (data8['dislikes'] > 0), data8['dislikes'] / data8['likes'], np.nan)
print(data8.loc[data8['dis/like'] == data8['dis/like'].min()]['title'].tolist()[0])

print("\n9. 가장많은 트렌드 영상을 제작한 채널의 이름은 무엇인가? (날짜기준, 중복포함)")
print(list(data.groupby(['channelId', 'channelTitle'], as_index=False).size().sort_values(by='size', ascending=False)['channelTitle'])[0])

print("\n10. 20회(20일)이상 인기동영상 리스트에 포함된 동영상의 숫자는?")
data10 = data.groupby(['title', 'channelId'], as_index=False).size()
print(len(data10.loc[data10['size'] >= 20]))
# print((data[['title','channelId']].value_counts()>=20).sum())
```

## 11~15 (*12, 15 dt-dt=Timedelta)

```python
# 데이터 출처 :https://www.kaggle.com/kukuroo3/youtube-episodic-contents-kr(참고, 데이터 수정)
# 데이터 설명 : 유튜브 “공범” 컨텐츠 동영상 정보 ( 10분 간격 수집)
DataUrl2_1 = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv' # (비디오 정보)
DataUrl2_2 = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv' # (참가자 채널 정보)
video = pd.read_csv(DataUrl2_1)
channel = pd.read_csv(DataUrl2_2)
display(video)
display(channel)
```

```python
import datetime
data15 = video.loc[video['videoname'].str.contains('EP1')].sort_values(by='ct')
data15['time_gap'] = data15['ct'].diff()
data15_id = data15.loc[(data15['time_gap'] <= datetime.timedelta(minutes=5)) | (data15['time_gap'] >= datetime.timedelta(minutes=20))].index.tolist()
d15_temp = []
for idx in data15_id:
    idm, idp = idx -1, idx + 1
    if idm not in data15_id:
        d15_temp.append(idm)
    if idp not in data15_id:
        d15_temp.append(idp)
data15_id += d15_temp
data15_id.sort()
display(data15.iloc[data15_id, :])
```

```python
print("11. 각 데이터의 'ct'컬럼을 시간으로 인식할수 있게 datatype을 변경하고 video 데이터의 videoname의 각 value 마다 몇개의 데이터씩 가지고 있는지 확인하라")
video['ct'] = pd.to_datetime(video['ct'])
channel['ct'] = pd.to_datetime(channel['ct'])
print("video['ct']:", video['ct'].dtype, "\nchannel['ct']:", channel['ct'].dtype)
display(video.groupby('videoname', as_index=False).size())

print("\n12. 수집된 각 video의 가장 최신화 된 날짜의 viewcount값을 출력하라")
display(video.sort_values(['videoname', 'ct']).drop_duplicates('videoname', keep='last')[['viewcnt', 'videoname', 'ct']])

print("\n13. Channel 데이터중 2021-10-03일 이후 각 채널의 처음 기록 됐던 구독자 수(subcnt)를 출력하라")
display(channel.loc[channel['ct'] >= pd.to_datetime('2021-10-03')].sort_values('ct').drop_duplicates('channelname', keep='first')[['channelname', 'subcnt']])

print("\n14. 각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라")
data14a = channel.loc[channel['ct'] >= pd.to_datetime('2021-10-03 03:00:00')].drop_duplicates('channelname', keep='first').sort_values(by='channelname').reset_index(drop=True)
data14b = channel.loc[channel['ct'] >= pd.to_datetime('2021-11-01 15:00:00')].drop_duplicates('channelname', keep='first').sort_values(by='channelname').reset_index(drop=True)
data14a = data14a.loc[:, ['channelname', 'subcnt']]
data14b = data14b.loc[:, ['channelname', 'subcnt']]
data14a.columns = ['channelname', 'subcnt_bf']
data14b.columns = ['channelname', 'subcnt_af']
data14c = pd.merge(data14a, data14b, how='inner')
data14c['subcnt_change'] = data14c['subcnt_af'] - data14c['subcnt_bf']
display(data14c.loc[:, ['channelname', 'subcnt_change']])

print("\n15. 각 비디오는 10분 간격으로 구독자수, 좋아요, 싫어요수, 댓글수가 수집된것으로 알려졌다.\
      공범 EP1의 비디오정보 데이터중 수집간격이 5분 이하, 20분이상인 데이터 구간( 해당 시점 전,후) 의 시각을 모두 출력하라")
import datetime

data15 = video.loc[video['videoname'].str.contains('EP1')].sort_values(by='ct')
data15['time_gap'] = data15['ct'].diff()
data15_id = data15.loc[(data15['time_gap'] <= datetime.timedelta(minutes=5)) | (data15['time_gap'] >= datetime.timedelta(minutes=20))].index.tolist()
d15_temp = []
for idx in data15_id:
    idm, idp = idx -1, idx + 1
    if idm not in data15_id:
        d15_temp.append(idm)
    if idp not in data15_id:
        d15_temp.append(idp)
data15_id += d15_temp
data15_id.sort()
display(data15.iloc[data15_id, :])
```

## 16~20 (*20 중복인 데이터 찾기)

```python
print("16. 각 에피소드의 시작날짜(년-월-일)를 에피소드 이름과 묶어 데이터 프레임으로 만들고 출력하라")
data16 = video.sort_values(by='ct').drop_duplicates('videoname', keep='first')[['ct', 'videoname']]
data16['ct'] = data16['ct'].dt.strftime("%Y-%m-%d")
display(data16)

print("\n17. '공범' 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다.\
      공개된 날의 21시의 viewcnt, ct, videoname 으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라")
display(video.loc[video['ct'].dt.hour == 21].drop_duplicates('videoname', keep='first').sort_values('viewcnt', ascending=False)[['viewcnt', 'ct', 'videoname']])

print("\n18. video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고\
      videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라")
video['ratio'] = video['dislikecnt'] / video['likecnt']
display(video.sort_values(by='ct', ascending=True).drop_duplicates('videoname', keep='last').sort_values(by='ratio', ascending=True)[['videoname', 'ratio']])

print("\n19. 2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오")
data19a = video.loc[video['ct'] >= pd.to_datetime('2021-11-01 00:00:00')].sort_values(by='ct').drop_duplicates('videoname', keep='first').sort_values(by='videoname')[['videoname', 'viewcnt']]
data19b = video.loc[video['ct'] <= pd.to_datetime('2021-11-01 15:00:00')].sort_values(by='ct').drop_duplicates('videoname', keep='last').sort_values(by='videoname')[['videoname', 'viewcnt']]
data19a.columns = ['videoname', 'viewcnt_bf']
data19b.columns = ['videoname', 'viewcnt_af']
data19c = pd.merge(data19a, data19b, how='inner')
data19c['viewcnt_change'] = data19c['viewcnt_af'] - data19c['viewcnt_bf']
display(data19c[['videoname', 'viewcnt_change']])
# def check(x):
#     result = max(x) - min(x)
#     return result
# answer = target[['videoname','viewcnt']].groupby("videoname").agg(check)

print("\n20. video 데이터 중에서 중복되는 데이터가 존재한다. 중복되는 각 데이터의 시간대와 videoname 을 구하여라")
display(video.loc[video.index.isin(set(video.index) - set(video.drop_duplicates().index))][['ct', 'videoname']])
```
