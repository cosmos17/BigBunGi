## 1~5 (*3)

```python
# ë°ì´í„° ì¶œì²˜ :https://www.kaggle.com/rsrishav/youtube-trending-video-dataset?select=KR_youtube_trending_data.csv
# ë°ì´í„° ì„¤ëª… : ìœ íŠœë¸Œ ë°ì¼ë¦¬ ì¸ê¸°ë™ì˜ìƒ (í•œêµ­)
DataUrl1 = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv'
data = pd.read_csv(DataUrl1, index_col=0)
display(data)
```

```python
print("1. ì¸ê¸°ë™ì˜ìƒ ì œì‘íšŸìˆ˜ê°€ ë§ì€ ì±„ë„ ìƒìœ„ 10ê°œëª…ì„ ì¶œë ¥í•˜ë¼ (ë‚ ì§œê¸°ì¤€, ì¤‘ë³µí¬í•¨)")
print(data.groupby(['channelTitle']).size().sort_values(ascending=False)[:10].index)

print("\n2. ë…¼ë€ìœ¼ë¡œ ì¸ê¸°ë™ì˜ìƒì´ ëœ ì¼€ì´ìŠ¤ë¥¼ í™•ì¸í•˜ê³  ì‹¶ë‹¤. dislikesìˆ˜ê°€ like ìˆ˜ë³´ë‹¤ ë†’ì€ ë™ì˜ìƒì„ ì œì‘í•œ ì±„ë„ì„ ëª¨ë‘ ì¶œë ¥í•˜ë¼")
print(data.loc[data['dislikes'] > data['likes']].groupby('channelTitle').size().index)

print("\n3. ì±„ë„ëª…ì„ ë°”ê¾¼ ì¼€ì´ìŠ¤ê°€ ìˆëŠ”ì§€ í™•ì¸í•˜ê³  ì‹¶ë‹¤. channelIdì˜ ê²½ìš° ê³ ìœ ê°’ì´ë¯€ë¡œ ì´ë¥¼ í†µí•´ ì±„ë„ëª…ì„ í•œë²ˆì´ë¼ë„ ë°”ê¾¼ ì±„ë„ì˜ ê°¯ìˆ˜ë¥¼ êµ¬í•˜ì—¬ë¼")
data3 = data[['channelId', 'channelTitle']].drop_duplicates()['channelId'].value_counts()
print(len(data3.loc[data3 > 1]))

print("\n4. ì¼ìš”ì¼ì— ì¸ê¸°ìˆì—ˆë˜ ì˜ìƒë“¤ì¤‘ ê°€ì¥ë§ì€ ì˜ìƒ ì¢…ë¥˜(categoryId)ëŠ” ë¬´ì—‡ì¸ê°€?")
data['trending_date2'] = pd.to_datetime(data['trending_date2'])
data4 = data.loc[data['trending_date2'].dt.weekday == 6]
print(data4.groupby('categoryId').size().sort_values().to_frame().index[-1])

print("\n5. ê° ìš”ì¼ë³„ ì¸ê¸° ì˜ìƒë“¤ì˜ categoryIdëŠ” ê°ê° ëª‡ê°œ ì”©ì¸ì§€ í•˜ë‚˜ì˜ ë°ì´í„° í”„ë ˆì„ìœ¼ë¡œ í‘œí˜„í•˜ë¼")
display(data.pivot_table(index=data['trending_date2'].dt.weekday, columns='categoryId', aggfunc='size'))
```

## 6~10 (*9)

```python
print("6. ëŒ“ê¸€ì˜ ìˆ˜ë¡œ (comment_count) ì˜ìƒ ë°˜ì‘ì— ëŒ€í•œ íŒë‹¨ì„ í•  ìˆ˜ ìˆë‹¤. viewcountëŒ€ë¹„ ëŒ“ê¸€ìˆ˜ê°€ ê°€ì¥ ë†’ì€ ì˜ìƒì„ í™•ì¸í•˜ë¼ (view_countê°’ì´ 0ì¸ ê²½ìš°ëŠ” ì œì™¸í•œë‹¤)")
data['ratio'] = np.where(data['view_count'] > 0, data['comment_count'] / data['view_count'], np.nan)
print(data.loc[data['ratio'] == data['ratio'].max()]['title'].tolist()[0])

print("\n7. ëŒ“ê¸€ì˜ ìˆ˜ë¡œ (comment_count) ì˜ìƒ ë°˜ì‘ì— ëŒ€í•œ íŒë‹¨ì„ í•  ìˆ˜ ìˆë‹¤.viewcountëŒ€ë¹„ ëŒ“ê¸€ìˆ˜ê°€ ê°€ì¥ ë‚®ì€ ì˜ìƒì„ í™•ì¸í•˜ë¼ (view_counts, ratioê°’ì´ 0ì¸ê²½ìš°ëŠ” ì œì™¸í•œë‹¤.)")
print(data.loc[data['ratio'] > 0].dropna().sort_values(by='ratio', ascending=True)['title'].iloc[0])

print("\n8. like ëŒ€ë¹„ dislikeì˜ ìˆ˜ê°€ ê°€ì¥ ì ì€ ì˜ìƒì€ ë¬´ì—‡ì¸ê°€? (like, dislike ê°’ì´ 0ì¸ê²½ìš°ëŠ” ì œì™¸í•œë‹¤)")
data8 = data.copy()
data8['dis/like'] = np.where((data8['likes'] > 0) & (data8['dislikes'] > 0), data8['dislikes'] / data8['likes'], np.nan)
print(data8.loc[data8['dis/like'] == data8['dis/like'].min()]['title'].tolist()[0])

print("\n9. ê°€ì¥ë§ì€ íŠ¸ë Œë“œ ì˜ìƒì„ ì œì‘í•œ ì±„ë„ì˜ ì´ë¦„ì€ ë¬´ì—‡ì¸ê°€? (ë‚ ì§œê¸°ì¤€, ì¤‘ë³µí¬í•¨)")
print(list(data.groupby(['channelId', 'channelTitle'], as_index=False).size().sort_values(by='size', ascending=False)['channelTitle'])[0])

print("\n10. 20íšŒ(20ì¼)ì´ìƒ ì¸ê¸°ë™ì˜ìƒ ë¦¬ìŠ¤íŠ¸ì— í¬í•¨ëœ ë™ì˜ìƒì˜ ìˆ«ìëŠ”?")
data10 = data.groupby(['title', 'channelId'], as_index=False).size()
print(len(data10.loc[data10['size'] >= 20]))
# print((data[['title','channelId']].value_counts()>=20).sum())
```

```python
# ë°ì´í„° ì¶œì²˜ :https://www.kaggle.com/kukuroo3/youtube-episodic-contents-kr(ì°¸ê³ , ë°ì´í„° ìˆ˜ì •)
# ë°ì´í„° ì„¤ëª… : ìœ íŠœë¸Œ â€œê³µë²”â€ ì»¨í…ì¸  ë™ì˜ìƒ ì •ë³´ ( 10ë¶„ ê°„ê²© ìˆ˜ì§‘)
DataUrl2_1 = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv' # (ë¹„ë””ì˜¤ ì •ë³´)
DataUrl2_2 = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv' # (ì°¸ê°€ì ì±„ë„ ì •ë³´)
video = pd.read_csv(DataUrl2_1)
channel = pd.read_csv(DataUrl2_2)
display(video)
display(channel)
```

```python
import datetime
data15 = video.loc[video['videoname'].str.contains('EP1')].sort_values(by='ct')
data15['time_gap'] = data15['ct'].diff()
data15_id = data15.loc[(data15['time_gap'] <= datetime.timedelta(minutes=5)) | (data15['time_gap'] >= datetime.timedelta(minutes=20))].index.tolist()
d15_temp = []
for idx in data15_id:
    idm, idp = idx -1, idx + 1
    if idm not in data15_id:
        d15_temp.append(idm)
    if idp not in data15_id:
        d15_temp.append(idp)
data15_id += d15_temp
data15_id.sort()
display(data15.iloc[data15_id, :])
```

```python
print("11. ê° ë°ì´í„°ì˜ 'ct'ì»¬ëŸ¼ì„ ì‹œê°„ìœ¼ë¡œ ì¸ì‹í• ìˆ˜ ìˆê²Œ datatypeì„ ë³€ê²½í•˜ê³  video ë°ì´í„°ì˜ videonameì˜ ê° value ë§ˆë‹¤ ëª‡ê°œì˜ ë°ì´í„°ì”© ê°€ì§€ê³  ìˆëŠ”ì§€ í™•ì¸í•˜ë¼")
video['ct'] = pd.to_datetime(video['ct'])
channel['ct'] = pd.to_datetime(channel['ct'])
print("video['ct']:", video['ct'].dtype, "\nchannel['ct']:", channel['ct'].dtype)
display(video.groupby('videoname', as_index=False).size())

print("\n12. ìˆ˜ì§‘ëœ ê° videoì˜ ê°€ì¥ ìµœì‹ í™” ëœ ë‚ ì§œì˜ viewcountê°’ì„ ì¶œë ¥í•˜ë¼")
display(video.sort_values(['videoname', 'ct']).drop_duplicates('videoname', keep='last')[['viewcnt', 'videoname', 'ct']])

print("\n13. Channel ë°ì´í„°ì¤‘ 2021-10-03ì¼ ì´í›„ ê° ì±„ë„ì˜ ì²˜ìŒ ê¸°ë¡ ëë˜ êµ¬ë…ì ìˆ˜(subcnt)ë¥¼ ì¶œë ¥í•˜ë¼")
display(channel.loc[channel['ct'] >= pd.to_datetime('2021-10-03')].sort_values('ct').drop_duplicates('channelname', keep='first')[['channelname', 'subcnt']])

print("\n14. ê°ì±„ë„ì˜ 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 ê¹Œì§€ êµ¬ë…ììˆ˜ (subcnt) ì˜ ì¦ê°€ëŸ‰ì„ êµ¬í•˜ì—¬ë¼")
data14a = channel.loc[channel['ct'] >= pd.to_datetime('2021-10-03 03:00:00')].drop_duplicates('channelname', keep='first').sort_values(by='channelname').reset_index(drop=True)
data14b = channel.loc[channel['ct'] >= pd.to_datetime('2021-11-01 15:00:00')].drop_duplicates('channelname', keep='first').sort_values(by='channelname').reset_index(drop=True)
data14a = data14a.loc[:, ['channelname', 'subcnt']]
data14b = data14b.loc[:, ['channelname', 'subcnt']]
data14a.columns = ['channelname', 'subcnt_bf']
data14b.columns = ['channelname', 'subcnt_af']
data14c = pd.merge(data14a, data14b, how='inner')
data14c['subcnt_change'] = data14c['subcnt_af'] - data14c['subcnt_bf']
display(data14c.loc[:, ['channelname', 'subcnt_change']])

print("\n15. ê° ë¹„ë””ì˜¤ëŠ” 10ë¶„ ê°„ê²©ìœ¼ë¡œ êµ¬ë…ììˆ˜, ì¢‹ì•„ìš”, ì‹«ì–´ìš”ìˆ˜, ëŒ“ê¸€ìˆ˜ê°€ ìˆ˜ì§‘ëœê²ƒìœ¼ë¡œ ì•Œë ¤ì¡Œë‹¤.\
      ê³µë²” EP1ì˜ ë¹„ë””ì˜¤ì •ë³´ ë°ì´í„°ì¤‘ ìˆ˜ì§‘ê°„ê²©ì´ 5ë¶„ ì´í•˜, 20ë¶„ì´ìƒì¸ ë°ì´í„° êµ¬ê°„( í•´ë‹¹ ì‹œì  ì „,í›„) ì˜ ì‹œê°ì„ ëª¨ë‘ ì¶œë ¥í•˜ë¼")
import datetime

data15 = video.loc[video['videoname'].str.contains('EP1')].sort_values(by='ct')
data15['time_gap'] = data15['ct'].diff()
data15_id = data15.loc[(data15['time_gap'] <= datetime.timedelta(minutes=5)) | (data15['time_gap'] >= datetime.timedelta(minutes=20))].index.tolist()
d15_temp = []
for idx in data15_id:
    idm, idp = idx -1, idx + 1
    if idm not in data15_id:
        d15_temp.append(idm)
    if idp not in data15_id:
        d15_temp.append(idp)
data15_id += d15_temp
data15_id.sort()
display(data15.iloc[data15_id, :])
```
